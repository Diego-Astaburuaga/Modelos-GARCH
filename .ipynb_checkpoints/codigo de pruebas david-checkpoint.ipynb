```python
from arch import arch_model
import arch.data.sp500

import datetime as dt
```

```python
st = dt.datetime(1988, 1, 1)
en = dt.datetime(2018, 1, 1)
data = arch.data.sp500.load()
market = data["Adj Close"]
returns = 100 * market.pct_change().dropna()
ax = returns.plot()
xlim = ax.set_xlim(returns.index.min(), returns.index.max())
```

# GARCH (with a Constant Mean)

```python
am = arch_model(returns)
res = am.fit(update_freq=5)
print(res.summary())
```

```python

fig = res.plot(annualize="D")
```

# GJR-GARCH

```python

am = arch_model(returns, p=3, o=0, q=3)
res = am.fit(update_freq=5, disp="off")
print(res.summary())
```

#  TARCH

```python

am = arch_model(returns, p=1, o=1, q=1, power=1.0)
res = am.fit(update_freq=5)
print(res.summary())
```

# Simulacin

```python
sim_mod = arch_model(None, p=1, o=0, q=0)
params = {"mu":0,"alpha[1]":2.1,"omega": 0.4}
param = pd.Series(params)


sim_data = sim_mod.simulate(param, 1000)
```
